```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

site= "https://forecast.weather.gov/MapClick.php?lat=37.77493000000004&lon=-122.41941999999995#.Y-zPYHZBxD8"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(site,headers=hdr)
page = urlopen(req)
data = BeautifulSoup(page, 'html.parser')
print(data)
```

```python
# find all the sections with specifiedd class name
div_data = data.find_all('li', attrs = {"class", "forecast-tombstone"})
```

```python
for i in div_data:
    print("Day = " + i.p.text)
    print("Discription = " + i.img['title'])
    weather = i.find('p', attrs = {"class", "short-desc"})
    print("weather = " + weather.text)
    temperature = i.find('p', attrs = {'class', 'temp'})
    print("temperature = " + temperature.text)
    print('\n')
```

```python
import pandas as pd


# create a list to store the data
scraped_data = []

for i in div_data:

    # initialize the dictionary
    Details = {}
    
    # Get data
    Day = i.p.text
    Title = i.img['title']
    weather = i.find('p', attrs = {"class", "short-desc"})
    weather = weather.text
    temperature = i.find('p', attrs = {'class', 'temp'})
    temperature = temperature.text
    
    # add data to the dictionary
    Details['Day'] = Day
    Details['Title'] = Title
    Details['Weather'] = weather
    Details['Temperature'] = temperature

    # append the scraped data to the list
    scraped_data.append(Details)
    
```

```python
# create a data frame from the list of dictionaries
dataFrame = pd.DataFrame.from_dict(scraped_data)

print(dataFrame)

# save the scraped data as CSV file
dataFrame.to_csv('C:\\Users\\manzo\\OneDrive\\Desktop\\Semester 08\\Data Science\\Webscrapping\\data.csv', index=False)
```
