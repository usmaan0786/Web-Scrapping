```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

site= "https://www.passiton.com/inspirational-quotes"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(site,headers=hdr)
page = urlopen(req)
data = BeautifulSoup(page, 'html.parser')
print(data)
```

```python
# find all the sections with specifiedd class name
div_data = data.find_all('div', attrs={"class","col-6 col-lg-4 text-center margin-30px-bottom sm-margin-30px-top"})
```

```python
print(div_data)
```

```python
print('Total Number of divs found = ', len(div_data))
```

```python
for i in div_data:
    
    quote = i.h5.a['href']
    quote_url = i.img['alt']
    image_link = i.img['src']
    category = i.h5.text
    
    print(quote)
    print(quote_url)
    print(image_link)
    print(category)
```

```python
import pandas as pd


# create a list to store the data
scraped_data = []

for i in div_data:

    # initialize the dictionary
    Details = {}

    # Get data
    quote = i.h5.a['href']
    quote_url = i.img['alt']
    image_link = i.img['src']
    category = i.h5.text
    
    # add data to the dictionary
    Details['quote'] = quote
    Details['quote_url'] = quote_url
    Details['image_link'] = image_link
    Details['category'] = category

    # append the scraped data to the list
    scraped_data.append(Details)

    
```

```python
# create a data frame from the list of dictionaries
dataFrame = pd.DataFrame.from_dict(scraped_data)

print(dataFrame)

# save the scraped data as CSV file
dataFrame.to_csv('C:\\Users\\manzo\\OneDrive\\Desktop\\Semester 08\\Data Science\\data.csv', index=False)
```
